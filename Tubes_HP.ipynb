```
import numpy as np 
import pandas as pd 

latih = pd.read_csv('https://drive.google.com/u/0/uc?id=1-F8zma3VS0QrX6iaHSFLYcEY_W2vKVB_&export=download')
tes = pd.read_csv('https://drive.google.com/uc?id=1BxUJMnGiLgTVEb-O0kkNEPubqDpy71bE&export=download')
sample_kirim= pd.read_csv('https://drive.google.com/uc?id=1hRwVCyiv4qP_5EN1nYVbiB04d34zPztX&export=download')
```

```
#cek missing value/ NaN
print(latih['Kelas Pekerja'].unique())
print(latih['Pekerjaan'].unique())
```

```
#ubah nilai ?
print(latih['Kelas Pekerja'].mode().values[0]) #modus Kelas pekerja: wiraswasta
latih['Kelas Pekerja'] = latih['Kelas Pekerja'].replace({'?': 'Wiraswasta'}) #ganti tiap nilai ? pd kolom kelas pekerja mjd wiraswasta
tes['Kelas Pekerja'] = tes['Kelas Pekerja'].replace({'?': 'Wiraswasta'})
```

```
latih['Pekerjaan'].value_counts()
```

```
#mengubah setiap kategori dengan num (umur)
umur_label= [1,2,3,4,5]
latih['Umur']=pd.qcut(latih['Umur'],5, labels=umur_label)
tes['Umur']=pd.qcut(tes['Umur'],5, labels=umur_label)
```

```
#mengubah setiap kategori dengan num (berat)
berat_label= [1,2,3,4,5]
latih['Berat Akhir']=pd.qcut(latih['Berat Akhir'],5, labels=berat_label)
tes['Berat Akhir']=pd.qcut(tes['Berat Akhir'],5, labels=berat_label)
```

```
#mengubah setiap kategori dengan num (kelas)
latih['Kelas Pekerja']=latih['Kelas Pekerja'].replace({'Wiraswasta':1, 'Pemerintah Lokal':2, 'Pekerja Bebas Perusahaan':3,
       'Pemerintah Negara':4, 'Pekerja Bebas Bukan Perusahan':5,
       'Pemerintah Provinsi':6, 'Tidak Pernah Bekerja':7, 'Tanpa di Bayar':8})
tes['Kelas Pekerja']=tes['Kelas Pekerja'].replace({'Wiraswasta':1, 'Pemerintah Lokal':2, 'Pekerja Bebas Perusahaan':3,
       'Pemerintah Negara':4, 'Pekerja Bebas Bukan Perusahan':5,
       'Pemerintah Provinsi':6, 'Tidak Pernah Bekerja':7, 'Tanpa di Bayar':8})
```

```
#Ubah Pendidikan
latih['Pendidikan']=latih['Pendidikan'].replace(
    {'SMA':1,'10th':2,'D3':3,'Sarjana':4,'Master':5,'Pendidikan Tinggi':6,
     '1st-4th':7,'Sekolah Professional':8,'7th-8th':9,'Doktor':10,
     '5th-6th':11,'11th':12,'9th':13,'D4':14, '12th':15,'SD':16})
tes['Pendidikan']=tes['Pendidikan'].replace(
    {'SMA':1,'10th':2,'D3':3,'Sarjana':4,'Master':5,'Pendidikan Tinggi':6,
     '1st-4th':7,'Sekolah Professional':8,'7th-8th':9,'Doktor':10,
     '5th-6th':11,'11th':12,'9th':13,'D4':14, '12th':15,'SD':16})
```

```
#Ubah Jenis Kelamin
latih['Jenis Kelamin']=latih['Jenis Kelamin'].replace({'Perempuan':1,'Laki2':2})
tes['Jenis Kelamin']=tes['Jenis Kelamin'].replace({'Perempuan':1,'Laki2':2})
```

```
#ubah status perkawinan
latih['Status Perkawinan'] = latih['Status Perkawinan'].replace(
    {'Menikah':1, 'Belum Pernah Menikah':2,'Cerai':3, 'Berpisah':4, 
     'Janda':5, 'Menikah LDR':6})
tes['Status Perkawinan'] = tes['Status Perkawinan'].replace(
    {'Menikah':1, 'Belum Pernah Menikah':2,'Cerai':3, 'Berpisah':4, 
     'Janda':5, 'Menikah LDR':6})
```

```
#Ubah Pekerjaan
latih['Pekerjaan']=latih['Pekerjaan'].replace({'Servis Lainnya':1, 'Ekesekutif Managerial':2, 'Spesialis':3,
       'Perbaikan Kerajinan':4, '?':5, 'Sales':6, 'Pembersih':7, 'Pemuka Agama':8,
       'Petani':9, 'Tech-support':10, 'Mesin Inspeksi':11, 'Supir':12,
       'Asisten Rumah Tangga':13, 'Penjaga':14, 'Tentara':15})
tes['Pekerjaan']=tes['Pekerjaan'].replace({'Servis Lainnya':1, 'Ekesekutif Managerial':2, 'Spesialis':3,
       'Perbaikan Kerajinan':4, '?':5, 'Sales':6, 'Pembersih':7, 'Pemuka Agama':8,
       'Petani':9, 'Tech-support':10, 'Mesin Inspeksi':11, 'Supir':12,
       'Asisten Rumah Tangga':13, 'Penjaga':14, 'Tentara':15})
```

```
#Ubah Keuntungan& Kerugian
a, untung =pd.cut(latih['Keuntungan Kapital'],5 ,labels=[1,2,3,4,5], retbins=True)
latih['Keuntungan Kapital']= pd.cut(latih['Keuntungan Kapital'], 
                                    bins=untung, labels=[1,2,3,4,5])
tes['Keuntungan Kapital']= pd.cut(tes['Keuntungan Kapital'], 
                                  bins=untung, labels=[1,2,3,4,5])

#Memberikan pembagian label pada kerugian capital
b, rugi = pd.cut(latih['Kerugian Capital'],5 ,labels=[1,2,3,4,5], retbins=True)
latih['Kerugian Capital']= pd.cut(latih['Kerugian Capital'],
                                  bins=rugi, labels=[1,2,3,4,5])
tes['Kerugian Capital']= pd.cut(tes['Kerugian Capital'], 
                                bins=rugi, labels=[1,2,3,4,5])
```

```
#Ubah jam kerja
c, jam = pd.cut(latih['Jam per Minggu'],5, labels=[1,2,3,4,5],retbins=True)
latih['Jam per Minggu']=pd.cut(latih['Jam per Minggu'],bins=jam, labels=[1,2,3,4,5])
tes['Jam per Minggu']=pd.cut(tes['Jam per Minggu'],bins=jam, labels=[1,2,3,4,5])
```

```
#Ubah Gaji di data train 
latih['Gaji']=latih['Gaji'].replace({'<=7jt':0,'>7jt':1})
```

```
#cek ada nilai NaN lagi 
print(latih.isnull().sum().sum())
print(tes.isnull().sum().sum())
```

```
latih_new=latih.drop(['id','Gaji'], axis=1).copy()
tes_new=tes.drop(['id'], axis=1).copy()
```

```
from sklearn.preprocessing import StandardScaler
stand=StandardScaler()
stand_l=stand.fit_transform(latih_new)
stand_t=stand.fit_transform(tes_new)
col_name=['Umur',	'Kelas Pekerja',	'Berat Akhir',	
          'Pendidikan',	'Jmlh Tahun Pendidikan',	'Status Perkawinan',
          'Pekerjaan',	'Jenis Kelamin',	'Keuntungan Kapital',	'Kerugian Capital',	
          'Jam per Minggu']
latih_std=pd.DataFrame(stand_l, columns=col_name) #dipake pemodelan
tes_std=pd.DataFrame(stand_t, columns=col_name) #dipake pemodelan
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import roc_auc_score

X_train = latih_std
Y_train = latih["Gaji"]
X_test  = tes_std
X_train.shape, Y_train.shape, X_test.shape
```

```
#Decision Tree
model_DT = DecisionTreeClassifier(criterion='gini',splitter='random')
model_DT.fit(X_train, Y_train)
Y_pred_DT = model_DT.predict(X_test)
model_DT.score(X_train, Y_train)
acc_model_DT = round(model_DT.score(X_train, Y_train) * 100, 2)
acc_model_DT
```

```
#Penilaian ROC AUC SCORE Decision Tree
Y_train_DT = model_DT.predict(X_train)
round(roc_auc_score(Y_train,Y_train_DT),2)*100
```
